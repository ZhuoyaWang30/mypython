```python
#Q1
import pandas as pd
```

```python
house = pd.read_csv('BostonHousing.csv')
attr = house.values
print house.columns.values
```

```python
#1.1
from matplotlib import pyplot as plt

plt.scatter(attr[:,-2], attr[:,4], color = '#1E9E47')

plt.xlabel('LSTAT')
plt.ylabel('NOX')

plt.show()
```

```python
#1.2
target = (attr[:,-1] > 20) == 1

plt.scatter(attr[:,-2], attr[:,4], c = target, alpha = 0.7)

plt.xlabel('LSTAT')
plt.ylabel('NOX')

plt.colorbar(ticks=[0,1])

plt.show()
```

```python
#1.3
target = (attr[:,-1] > 20) 

larger = attr[target == 1, :]
smaller = attr[target == 0, :]
num_bins = 10
plt.hist(larger[:, 1], num_bins, alpha =1)
plt.hist(smaller[:, 1], num_bins, alpha = 0.8)

plt.xlabel('ZN')
plt.ylabel('Frequency')

plt.legend(['MEDV>20000', 'MEDV<=20000'])

plt.show()
```

```python
#1.4
target = (attr[:,-1] > 20) 

larger = attr[target == 1, :]
smaller = attr[target == 0, :]
num_bins = 10
plt.hist(larger[:, -2], num_bins, alpha =1)
plt.hist(smaller[:, -2], num_bins, alpha = 0.8)

plt.xlabel('LSTAT')
plt.ylabel('Frequency')

plt.legend(['MEDV>20000', 'MEDV<=20000'])

plt.show()
```

```python
#Q2
import pandas as pd
appliance = pd.read_csv('ApplianceShipments.csv')
attr = appliance.values
print appliance.columns.values
```

```python
import matplotlib.pyplot as plt
#2.1

x = attr[0:4,1]
y = attr[0:4,2]

plt.plot(x,y,linewidth = 4)

plt.xlabel('Quarter')
plt.ylabel('Shipments(in millions of dollars)')

plt.show()
```

```python
#2.2
target_1 = attr[:,1] == 1
target_2 = attr[:,1] == 2
target_3 = attr[:,1] == 3
target_4 = attr[:,1] == 4

x_1 = attr[target_1,0]
x_2 = attr[target_2,0]
x_3 = attr[target_3,0]
x_4 = attr[target_4,0]
Q_1 = attr[target_1,2]
Q_2 = attr[target_2,2]
Q_3 = attr[target_3,2]
Q_4 = attr[target_4,2]

plt.plot(x_1,Q_1)
plt.plot(x_2,Q_2)
plt.plot(x_3,Q_3)
plt.plot(x_4,Q_4)

plt.xlabel('Year')
plt.ylabel('Shipments over years (in millions of dollars)')

plt.legend(['Q1','Q2','Q3','Q4'])

plt.show()
```

```python
#2.3
import numpy as np
target_1 = attr[:,0] == 1985
target_2 = attr[:,0] == 1986
target_3 = attr[:,0] == 1987
target_4 = attr[:,0] == 1988
target_5 = attr[:,0] == 1989
xx = [1985,1986,1987,1988,1989]
y_1 = np.sum(attr[target_1,2])
y_2 = np.sum(attr[target_2,2])
y_3 = np.sum(attr[target_3,2])
y_4 = np.sum(attr[target_4,2])
y_5 = np.sum(attr[target_5,2])
y_sum = [y_1,y_2,y_3,y_4,y_5]
```
